# 1. Import libraries

```python
import pandas as pd
from sqlalchemy import create_engine
```

# 2. Import datasets

```python
# message

messages_df = pd.read_csv(r'disaster_messages.csv')
messages_df.head()
```

```python
# categories

categories_df = pd.read_csv(r'disaster_categories.csv')
categories_df.head()
```

```python
# Merge datasets

df = messages_df.merge(categories_df, left_on='id', right_on='id')
df.head()
```

# 3. Clean data

```python
# Create a dataframe of the 36 individual category columns

categories = df['categories'].str.split(';', expand=True)
categories.head()
```

```python
# Get the names for 36 columns

row = categories.iloc[0,:]
category_colnames = row.apply(lambda x: x.partition('-')[0])

categories.columns = category_colnames
categories.head()
```

```python
# Convert values in 'categories' dataframe in to 1 and 0

for column in categories.columns:
    categories[column] = categories[column].apply(lambda x: x.partition('-')[2])
    categories[column] = categories[column].astype('int32')
```

```python
# Concat 'df' dataframe with the new 'categories' dataframe

df.drop('categories', axis=1, inplace=True)

df = pd.concat([df, categories], axis=1)
df.head()
```

```python
# Drop duplicates

df.drop_duplicates(inplace=True)
```

# 4. Export to an sqlite database

```python
# Export to an sqlite database

engine = create_engine('sqlite:///DisasterResponse.db')
df.to_sql('DisasterResponse', engine, index=False)
```
